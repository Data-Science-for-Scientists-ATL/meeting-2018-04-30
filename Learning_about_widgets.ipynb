$A = amplitude$

$\phi = phase$

$\nu = freq$

$A \sin(2 \pi(\nu x+\phi))$

```python
#define x variable - from zero to 2 with 1000 points
x=np.linspace(0,2,1000)
#set up figure and axes - one subplot, give figure size
fig, ax=plt.subplots(1,figsize=(10,4))
#give a title to the plot
plt.suptitle('Sin Wave')

#define a function and your inputs
def update_plot(amp, phase, freq):
    
    
    """
    This text is a 'docstring' which describes your function
    This function is linked to the sliders and
    it replots the sine wave when the sliders are changed
    """
    
    #You need to clear the axes each time so you don't have waves on waves on waves
    ax.clear()
    
    #define label that goes into the legend
    units = 'amp = {} $(psi)$ \nphase = {} $(s)$ \nfreq = {} $(Hz)$'
    #define y function - see function we are plotting above
    y=amp *np.sin(2 * np.pi* (freq * x + phase))
    
    #define what you are plotting - give it the legend information
    ax.plot(x,y, label=units.format(amp, phase, freq))
    ax.legend(loc='upper right')
    ax.set_xlabel('$(s)$')
    ax.set_ylabel('$(psi)$')
    ax.set_xlim(x[0],x[-1])
    plt.show()
    

#define your widgets - what your sliders are
#float sliders give you decimal points - default is two dp, horizontal
amp=widgets.FloatSlider(min=1, max=10, value=1, description='Amp:')
phase=widgets.FloatSlider(min=0, max=5, value=0, description='Phase:')
freq=widgets.FloatSlider(min=1, max=10, value=1, description='Freq:')

#call widgets.interactive to call the function and give it your widgets as the arguments
#links the value of the widgets as the argument of the function for the plot
widgets.interactive(update_plot, amp=amp, phase=phase, freq=freq)
```
